```python
"""
    Code by Tae-Hwan Hung(@graykode)
    https://en.wikipedia.org/wiki/Gamma_distribution
"""
import numpy as np
from matplotlib import pyplot as plt

# Calculate the gamma function for a given n
def gamma_function(n):
    cal = 1
    for i in range(2, n):
        cal *= i
    return cal

# Calculate gamma distribution values for given x, alpha (a) and beta (b) parameters
def gamma(x, a, b):
    # Calculate the normalization constant c
    c = (b ** a) / gamma_function(a)
    # Calculate gamma distribution values
    y = c * (x ** (a - 1)) * np.exp(-b * x)
    return x, y, np.mean(y), np.std(y)

# Plot gamma distributions for different alpha and beta parameters
for ls in [(1, 1), (2, 1), (3, 1), (2, 2)]:
    a, b = ls[0], ls[1]

    # Generate x values from 0 to 20
    x = np.arange(0, 20, 0.01, dtype=float)
    # Calculate gamma distribution values
    x, y, u, s = gamma(x, a=a, b=b)
    # Plot the distribution with mean, std, alpha and beta in legend
    plt.plot(x, y, label=r'$\mu=%.2f,\ \sigma=%.2f,'
                         r'\ \alpha=%d,\ \beta=%d$' % (u, s, a, b))
plt.legend()
plt.savefig('graph/gamma.png')
plt.show()
```
